## Phương thức shift()

```python
ACB = ACB.to_frame()  # Chuyển đổi biến ACB từ Series thành DataFrame
```

```python
ACB.columns = ['Giá hôm nay']  # Đổi tên cột duy nhất trong DataFrame ACB thành 'Giá hôm nay'
```

```python
ACB.head(6)  # Hiển thị 6 dòng đầu tiên của DataFrame ACB
```

```python
ACB.shift(periods = 1) # Dịch chuyển dữ liệu xuống 1 dòng, giữ nguyên chỉ mục
```

```python
ACB['Giá hôm trước'] = ACB.shift(periods=1)  # Thêm cột 'Giá hôm trước' chứa giá của ngày trước đó
```

```python
ACB
```

```python
ACB['Biến động'] = ACB['Giá hôm nay'] - ACB['Giá hôm trước']  # Tính mức biến động giá so với ngày trước đó
```

```python
ACB
```

```python
ACB['TL biến động'] = round(ACB['Giá hôm nay'].div(ACB['Giá hôm trước']).sub(1).mul(100), 2)
# Tính tỷ lệ biến động (%) của giá cổ phiếu so với ngày trước đó và làm tròn đến 2 chữ số thập phân
```

```python
ACB
```

## Phương thức diff() và pct_change()

```python
round(ACB['Giá hôm nay'].pct_change(periods=1).mul(100), 2)
# Tính phần trăm biến động giá so với ngày trước đó và làm tròn đến 2 chữ số thập phân
```

## Đánh giá lợi nhuận, rủi ro giữa các mã cổ phiếu

```python
loinhuan_vn30 = giadongcua.pct_change().dropna()  
# Tính tỷ lệ biến động giá hàng ngày (lợi nhuận) của tất cả cổ phiếu trong DataFrame 'giadongcua' và loại bỏ giá trị NaN
```

```python
loinhuan_vn30.head()
```

```python
loinhuan_vn30.describe()
# Thống kê mô tả về tỷ lệ biến động hàng ngày của tất cả cổ phiếu trong DataFrame 'loinhuan_vn30'
```

```python
tomtat_vn30 = loinhuan_vn30.describe().T.loc[:, ["mean", "std"]]  
# Trích xuất chỉ giá trị trung bình ('mean') và độ lệch chuẩn ('std') từ thống kê mô tả của 'loinhuan_vn30'
```

```python
tomtat_vn30["mean"] = tomtat_vn30["mean"] * 250
# Chuyển đổi lợi nhuận trung bình từ hàng ngày sang hàng năm bằng cách nhân với 249 ngày giao dịch

tomtat_vn30["std"] = tomtat_vn30["std"] * np.sqrt(250)  
# Chuyển đổi độ lệch chuẩn từ hàng ngày sang hàng năm bằng cách nhân với căn bậc hai của 249
```

```python
tomtat_vn30
```

```python
# Vẽ biểu đồ scatter với màu sắc đẹp hơn và kích thước điểm dễ nhìn hơn
plt.figure(figsize=(13, 7))
plt.scatter(tomtat_vn30["std"], tomtat_vn30["mean"], s=50, c="blue", alpha=0.7, edgecolors="black")

# Gắn nhãn tên cổ phiếu
for i in tomtat_vn30.index:
    plt.annotate(i, 
                 xy=(tomtat_vn30.loc[i, "std"], tomtat_vn30.loc[i, "mean"]), 
                 xytext=(5, 5), textcoords='offset points', 
                 fontsize=12, color="darkred")

# Đánh dấu cổ phiếu có lợi nhuận cao nhất
best_stock = tomtat_vn30.loc[tomtat_vn30["mean"].idxmax()]  # Cổ phiếu có lợi nhuận cao nhất
plt.scatter(best_stock["std"], best_stock["mean"], s=120, c="red", edgecolors="black", label="Lợi nhuận cao nhất")

# Thêm đường gốc (0,0) để dễ so sánh rủi ro & lợi nhuận
plt.axhline(y=0, color='black', linestyle='dashed', linewidth=1)  # Đường lợi nhuận = 0
plt.axvline(x=0, color='black', linestyle='dashed', linewidth=1)  # Đường rủi ro = 0

# Thêm tiêu đề & nhãn trục
plt.xlabel("Biến động năm (Độ lệch chuẩn)", fontsize=15)
plt.ylabel("Lợi nhuận năm (%)", fontsize=15)
plt.title("Lợi nhuận & Biến động của cổ phiếu năm 2024", fontsize=18, fontweight="bold")

# Hiển thị chú thích và lưới nhẹ để dễ quan sát
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

# Lưu biểu đồ
plt.savefig("loinhuan_biendong.png", dpi=300) 
plt.show()
```
